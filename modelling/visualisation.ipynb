```python
from sklearn.manifold import TSNE

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme()
```

```python
df = pd.read_csv("precontrast_pyradiomics.csv")
```

```python
features = df.drop(columns = ['patient', 'sequence', 'ER', 'PR', 'HER2', 'Mol Subtype'])
```

```python
tsne = TSNE(2, perplexity = 50, verbose = 2, n_jobs = -1)
tsne_feats = tsne.fit_transform(features)
```

```python
tsne_df = pd.DataFrame(tsne_feats, columns = ['X', 'Y'])

tsne_df[['ER', 'PR', 'HER2', 'Mol Subtype']] = df[['ER', 'PR', 'HER2', 'Mol Subtype']]
```

## ER

```python
plt.figure(figsize = (16, 8))
sns.scatterplot(x = 'X', y = 'Y', hue = 'ER', data = tsne_df)

plt.xlabel("X")
plt.ylabel("Y")
plt.title("Visualisation for ER")
```

## PR

```python
plt.figure(figsize = (16, 8))
sns.scatterplot(x = 'X', y = 'Y', hue = 'PR', data = tsne_df)

plt.xlabel("X")
plt.ylabel("Y")
plt.title("Visualisation for PR")
```

## HER2

```python
plt.figure(figsize = (16, 8))
sns.scatterplot(x = 'X', y = 'Y', hue = 'HER2', data = tsne_df)

plt.xlabel("X")
plt.ylabel("Y")
plt.title("Visualisation for HER2")
```

## Molecular Subtype

```python
plt.figure(figsize = (16, 8))
sns.scatterplot(x = 'X', y = 'Y', hue = 'Mol Subtype', data = tsne_df)

plt.xlabel("X")
plt.ylabel("Y")
plt.title("Visualisation for Molecular Subtype")
```
